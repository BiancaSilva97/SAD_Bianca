    1) Aggregation
    2) Sampling

    5) Feature Subset Selection 
    6) Feature Creation

    8) Attribute Transformation


O que é feito em cada caso, é da sua inteira liberdade. 




````python
import sklearn.datasets as datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = datasets.california_housing.fetch_california_housing()
dframe = pd.DataFrame(data = data.data, columns=data.feature_names)

dframe.head()

````

````python
#Aggregation
ag=dframe.agg(["min","max", "mean"])
ag
````

````python
#Sampling
sp=dframe.sample(5)
sp
````

````python
#Dimensionality Reduction
from sklearn.decomposition import PCA
pca = PCA(n_components=4)
pca.fit(dframe)
columns = ['pca_%i' % i for i in range(4)]
df_pca = pd.DataFrame(pca.transform(dframe), columns=columns, index=dframe.index)
df_pca.head()
````

````python
#Feature Subset Selection 
fss=dframe[['AveRooms','AveBedrms','AveOccup']][0:10]
fss
````

````python
#Feature Creation
def calculate(x):
  return x+2
dframe["NewAge"] = calculate(dframe['HouseAge'])
dframe.head()
````

````python
#Discretization and Binarization
bn = dframe.HouseAge[0:10]
pd.get_dummies(bn)
````

````python
#Attribute Transformation
dframe['MeanAveOccup'] = dframe['AveOccup'].transform(lambda x : x/x.mean())
dframe.head()
````
